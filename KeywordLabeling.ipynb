```python
import os
import time
import pyaudio
import librosa
import IPython.display as ipd
AUDIO_FILE_PATH = os.path.join(os.getcwd(),"./data/Models/KeywordModel/Training/Sequences/Audio")
```

```python
def play_and_label_audio_sequence(audio_file_path):
    # Load audio sequence
    audio_sequence, sr = librosa.load(audio_file_path, sr=None, mono=True)
    
    # Compute the length of audio sequence in seconds
    audio_sequence_length = librosa.get_duration(audio_sequence, sr=sr)
    
    # Define the length of audio clip for labeling
    audio_clip_length = 0.25
    
    # Loop through audio sequence and play it for user
    for i in range(0, int(audio_sequence_length/audio_clip_length)):
        audio_clip = audio_sequence[int(i*sr*audio_clip_length):int((i+1)*sr*audio_clip_length)]
        print("Playing audio clip {}...".format(i+1))
        ipd.display(ipd.Audio(audio_clip, rate=sr))
        
        # Prompt user for label
        label = None
        while True:
            try:
                label = int(input("Enter the label for audio clip {} (0 or 1): ".format(i+1)))
                break
            except:
                continue
        # Get the file name from the audio_file_path
        file_name = os.path.basename(audio_file_path)
        # Store label in a file (you can modify this to store the labels in a database or CSV file)
        with open("labels.txt", "a") as f:
            f.write("{}\t{}\n".format(file_name, label))
```

```python
def get_remaining_audio_files(path):
    # Define path to directory with audio sequences
    dir_path = path

    # Get list of files in directory
    dir_files = os.listdir(dir_path)

    # Define path to labels file
    labels_file = 'labels.txt'
    labeled_files = []
    # Open and read labels file
    with open(labels_file, 'r') as f:
        for line in f:
            audio_file = line.strip().split('\t')[0]
            if audio_file not in labeled_files:
                labeled_files.append(audio_file)

    # Loop through labels and remove any matching strings from dir_files
    for label in labeled_files:
        if label in dir_files:
            print('Removing file: ' + label)
            dir_files.remove(label)

    # Print remaining files in dir_files
    print('Remaining Files:\n', dir_files)
    return dir_files
```

```python
def main():
    # Get a list of audio files in the folder
    audio_files = get_remaining_audio_files(AUDIO_FILE_PATH)
    # Loop through each audio file and play it for user
    for audio_file_path in audio_files:
        print("Playing audio file: {}".format(audio_file_path))
        audio_sequence, sr = librosa.load(os.path.join(AUDIO_FILE_PATH, audio_file_path), sr=None, mono=True)
        ipd.display(ipd.Audio(audio_sequence, rate=sr))
        play_and_label_audio_sequence(os.path.join(AUDIO_FILE_PATH, audio_file_path))
```

```python
main()
```
